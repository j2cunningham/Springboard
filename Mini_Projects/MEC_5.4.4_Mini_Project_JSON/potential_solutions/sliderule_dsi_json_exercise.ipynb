Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
```

```python
# Load json file as a string
data = json.load((open('data/world_bank_projects.json')))

# Load json file into a Pandas dataframe
data_df = pd.read_json('data/world_bank_projects.json')
```

```python
# 1. The 10 countries with the most projects

# First, we use the Pandas dataframe to isolate columns: country name and project names
top_projects = data_df[['countryname' , 'project_name']]

# Then, we group by country name 
top_projects = top_projects.groupby("countryname")

# Now, we get a count using .size() and we sort from highest to lowest using .sort_values()
top_projects = top_projects.size().sort_values(ascending = False)

#Finally, we use .head() to get top 10.
top_projects = top_projects.head(10)
top_projects
```

```python
# 2. The top 10 major project themes (using column 'mjtheme_namecode')

# First, we normalize the data to create tables from nested element.
top_project_themes = json_normalize(data, 'mjtheme_namecode')

# Then, we group by name using .groupby()

top_project_themes = top_project_themes.groupby('name')

# Now, we get a count using .size() and we sort from highest to lowest using .sort_values()
top_project_themes = top_project_themes.size().sort_values(ascending = False)

#Finally, we use .head() to get top 10.
top_project_themes = top_project_themes.head(10)
top_project_themes
```

```python
# 3. Notice that some entries have only the code and the name is missing. 
# Our objective is to create a dataframe with the missing names filled in.

# See that values with index 1, 13, 17, 19 and etc. in mjtheme_namecode are missing.
all_x = json_normalize(data, 'mjtheme_namecode')
all_x
```

```python
# Let's drop duplicate rows using .drop_duplicates to have a smaller table to work with.
some_x = all_x.drop_duplicates()
some_x
```

```python
# Now, we need to get rid of the empty rows then sort.

# First we replace the empty space as NA values.
some_x = some_x[some_x != '']

# Then, we drop all rows that have 'name' as NA
some_x.dropna(subset=['name'], inplace=True)

# We change 'code' to be an integer instead of a string so we can sort
x1 = some_x[['code']].astype(int)
x2 = some_x[['name']]

# Finally we combine using concat(), sort using sort_values(), and reset the index
x = pd.concat([x1, x2], axis=1)
x = x.sort_values("code")
x = x.reset_index(drop=True)
x
```
